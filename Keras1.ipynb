```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels=[]
train_samples=[]
```

Generating Random Data using randInt

1-An example data was tested on people aged between 13-100
2-the trial had 2100 people,where half are of younger age is (13-64) and other half are of older age falls in 65-100
3-95% of patients under older age had experienced side effects.
4-95% of patients under younger age have not experienced any side effects.

Here label would be (having a side effect/not having ) and samples would be ages of 2100 people.

```python
for i in range(1000):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)

for i in range(50):
  random_younger=randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)

  random_younger=randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)

#printing raw samples data    
for i in train_samples:
  print(i)
```

```python
#printing raw labels data
for i in train_labels:
  print(i)


print(train_samples)
```

```python
#converting samples and label data into numpy arrays since keras only accepts numpy arrays.
numpy_train_labels = np.array(train_labels)
numpy_train_samples = np.array(train_samples)

print(numpy_train_samples)
```

```python
#Now this data can be fed as input to keras,but however since the range is high between (13-100)the net may not learn efficiently,
#hence we need to scale it down within a range 0-1
scalar = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scalar.fit_transform(numpy_train_samples)
```

```python
for i in scaled_train_samples:
 print(i)
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model=Sequential([
    Dense(16,input_shape=(1,),activation='relu'),
    Dense(32,activation='relu'),
    Dense(2,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#train the model on training samples
#try removing verbose and check(look thru keras documentation for optional parameters and the meaning of all the parameters)
model.fit(scaled_train_samples,train_labels,validation_split=0.1,batch_size=10,epochs=20,shuffle=True,verbose=2)
```

```python
#Prepare Testing data

test_labels=[]
test_samples=[]

for i in range(10):
  test_random_younger = randint(13,64)
  test_samples.append(test_random_younger)
  train_labels.append(0)

  test_random_older = randint(65,100)
  test_samples.append(test_random_older)
  test_labels.append(1)

for i in range(200):
  test_random_younger=randint(13,64)
  test_samples.append(test_random_younger)
  test_labels.append(1)

  test_random_younger=randint(65,100)
  test_samples.append(test_random_younger)
  test_labels.append(0)
    
    
numpy_test_labels = np.array(test_labels)
numpy_test_samples = np.array(test_samples)
scaled_test_samples = scalar.fit_transform(numpy_test_samples)

print(scaled_test_samples)
```

```python
predictions=model.predict(scaled_test_samples,batch_size=10,verbose=0)

for i in predictions:
 print(i)
```

```python
rounded_predictions=model.predict_classes(scaled_test_samples,batch_size=10,verbose=0)

for i in rounded_predictions:
 print(i)
```
